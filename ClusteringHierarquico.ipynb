Importando os componentes

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.datasets import make_blobs
```

1. Gerando dados de exemplo

```python
x, y = make_blobs(n_samples = 100, centers = 3)
```

2. Criando a matriz de ligação (linkage matrix)

Método "Ward" minimiza a variação dentro dos clusters

```python
z = linkage (x, method ='ward')
```

3. "Plotando" o dendograma


```python
plt.figure(figsize = (10, 5))
dendrogram(z)
plt.title('Dendograma - Hierarchical Clustering')
plt.xlabel('Observações')
plt.ylabel('Distância')
plt.show()
```

4. Extraindo os clusters a partir do corte no dendograma

Critério 'maxclust' define o número de clusters desejado

```python
clusters = fcluster(z, t = 3, criterion = 'maxclust')
```

5. Visualizando os clusters resultantes

```python
plt.scatter(x[:,0], x[:,1], c = clusters, cmap = 'rainbow')
plt.title('Agrupamento Hierárquico (3 Clusters)')
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.show()
```
